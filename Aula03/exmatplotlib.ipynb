```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gera números com a Numpy
dados_x = np.linspace(0,10,10)    # gera valores com a biblioteca numpy
dados_y = dados_x * dados_x    # gera valores multiplicando um dado pelo outro  
```

```python
plt.figure(figsize=(10,5), facecolor="gray")
Ax = plt.axes() # cria os eixos do gráfico
Ax.set_facecolor('white') # cria o fundo do gráfico
plt.plot(dados_x, dados_y, color='red', linewidth=2, marker='o', linestyle='dashed', markeredgecolor='black')
plt.title("Exemplo Matplotlib")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
```

```python
plt.rcParams.update({'font.size':15})   # define o tamanho da fonte 
plt.figure(figsize=(12,12)) # define o tamanho da figura
linhas = 2    # define o número de linhas 
colunas = 2    # define o número de colunas
plt.subplot(linhas, colunas, 1)
plt.plot(dados_x, 'r--')
plt.title("Gráfico 1")

# gráfico 2
plt.subplot(linhas, colunas, 2)
plt.plot(dados_x*dados_x, 'b-.')
plt.title("Gráfico 2")

# gráfico 3
plt.subplot(linhas, colunas, 3)
plt.plot(dados_x/dados_x, 'c-.')
plt.title("Gráfico 3")

# gráfico 4
plt.subplot(linhas, colunas, 4)
plt.plot(dados_y*2)
plt.title("Gráfico 4")

plt.savefig("Exemplo subplot.jpg", dpi=300, transparent=True,
            bbox_inches='tight')    # comando para salvar a figura
 
```

```python
# gráfico de barras
figura = plt.figure()   # armazena na variável Figura a figura a ser criada
grafico = figura.add_axes([0, 0, 0.7, 0.7])     # cria a grade do gráfico
rotulos = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
grafico.bar(rotulos, dados_x, color='pink')# cachorro salsicha
grafico.bar(rotulos, dados_x*0.1, color='cyan')
plt.title("Quantidade X Categoria")
plt.xlabel("Itens")
plt.ylabel("Qualidade")






```
