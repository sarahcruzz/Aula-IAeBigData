```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
plt.figure(figsize=(6,6))
plt.boxplot(df['pressão'])

# Título e rótulos
plt.title('Boxplot da Pressão', fontsize=15)
plt.ylabel('Pressão')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
