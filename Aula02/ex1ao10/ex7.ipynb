```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python

# Criando o boxplot da coluna 'humidade' do dataframe 'df'
plt.figure(figsize=(6,6))
plt.boxplot(df['direção do vento'])

# Título e rótulos
plt.title('Boxplot da Direção do Vento', fontsize=15)
plt.ylabel('Direção do vento (Km/h)')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
