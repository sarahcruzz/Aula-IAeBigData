```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df['data'] = pd.to_datetime(df['data'])  
```

```python

max_pressao = df['pressão'].max()  # Valor máximo de umidade
min_pressao = df['pressão'].min()  # Valor mínimo de umidade


max_data = df['data'][df['humidade'].idxmax()]
min_data = df['data'][df['humidade'].idxmin()]

```

```python
plt.figure(figsize=(10,6)) 
plt.plot( df['data'],df['pressão'], color="purple") 
plt.title('Gráfico de Pressão', fontsize=15)

plt.annotate(f'Máximo: {max_pressao}%', xy=(max_data, max_pressao), xytext=(max_data, max_pressao), fontsize=12, color='red')
plt.annotate(f'Mínimo: {min_pressao}%', xy=(min_data, min_pressao), xytext=(min_data, min_pressao), fontsize=12, color='green')

plt.xlabel('Ano')
plt.ylabel('Pressão')
```
