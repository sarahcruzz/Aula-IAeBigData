```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
# Criando o boxplot da coluna 'humidade' do dataframe 'df'
plt.figure(figsize=(6,6))
plt.boxplot(df['humidade'])

# Título e rótulos
plt.title('Boxplot da Umidade', fontsize=15)
plt.ylabel('Umidade (%)')

# Exibindo o gráfico
plt.grid(True)
plt.show()
```
