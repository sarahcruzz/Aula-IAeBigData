```python
import pandas as pd
```

```python
df = pd.read_csv('imigrantes_canada.csv') # importa a base de dados
```

```python
df.head() # exibe as 4 primeiras linhas do dataframe
```

```python
df.info() # exibe as informações do dataframe
```

```python
df.set_index(['País'], inplace=True) # seta o index com a coluna País
anos = list(map(str, range(1980,2013))) # converte os anos em string armazenados em uma lista
```

```python
brasil = df.loc['Brasil', anos] # pega os dados somente para o país brasil
```

```python
argentina = df.loc['Argentina', anos]
```

```python
argentina
```

```python
brasil
```

```python
# cria um dicionário com os dados brasil
brasil_dic = {
    'ano':brasil.index.tolist(), # transforma a coluna com o index em uma lista
    'imigrantes':brasil.values.tolist() # transforma a coluna com os valores em uma lista
}
```

```python
argentina_dic = {
    'ano':argentina.index.tolist(), # transforma a coluna com o index em uma lista
    'imigrantes':argentina.values.tolist() # transforma a coluna com os valores em uma lista
}
```

```python
df_brasil = pd.DataFrame(brasil_dic) # cria o dataframe
```

```python
df_argentina = pd.DataFrame(argentina_dic)
```

```python
df_brasil # exibe os valores
```
