```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
lojas = ['A', 'B', 'C', 'D']
vendas_2022 = {'Jan': [100, 80, 150, 50],
               'Fev': [120, 90, 170, 60], 
               'Mar': [150, 100, 200, 80], 
               'Abr': [180, 110, 230, 90],
               'Mai': [220, 190, 350, 200],
               'Jun': [230, 150, 280, 120],
               'Jul': [250, 170, 300, 140],
               'Ago': [260, 180, 310, 150],
               'Set': [240, 160, 290, 130],
               'Out': [220, 140, 270, 110],
               'Nov': [400, 220, 350, 190],
               'Dez': [300, 350, 400, 250]}

# Criando o DataFrame a partir do dicionrio de vendas
df_vendas = pd.DataFrame(vendas_2022, index=lojas).T

# Criando a figura com subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Ttulos das lojas para os subplots
titulos = ['Loja A', 'Loja B', 'Loja C', 'Loja D']

# Iterando sobre as lojas para criar os grficos de linhas
for i, ax in enumerate(axs.flat):
    loja = lojas[i]
    ax.plot(df_vendas.index, df_vendas[loja], marker='o')
    ax.set_title(titulos[i])
    ax.set_xlabel('Meses')
    ax.set_ylabel('Vendas')
    ax.grid(True)

# Ajustando o layout da figura
plt.tight_layout()
plt.show()
```
