```python
import numpy as np # biblioteca numpy para
# fazer cálculos matemáticos
from sklearn.model_selection  import train_test_split # importa da biblioteca
#sklearn.model  a biblioteca train test split
# essa biblioteca train test split serve para  dividir os dados
# teste e treino
from sklearn.tree import DecisionTreeClassifier # vai importar o
#algoritmo  da arvore de decisão
from sklearn import tree # importa a biblioteca para plotar a arvore de decisão
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import seaborn as sns # importa a biblioteca seaborn
from sklearn.metrics import accuracy_score # métrica de acurácia do algoritmo
from sklearn.metrics import precision_score # métrica de precisao do algoritmo
from sklearn.metrics import recall_score # métrica de recall do algoritmo
from sklearn.metrics import f1_score # métrica de f1 score
from sklearn.metrics import confusion_matrix # importa a matriz de confusão
```

```python
#pip install scikit-learn 
```

```python
# criando o conjunto de dados
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])
classes = np.array(['bom','ruim','ruim','bom','pessimo',
                    'pessimo','bom','pessimo',
                     'pessimo','ruim','ruim','bom','bom'])
```

```python
#dividir dados em teste e treino
notas_treino, notas_teste, classes_treino,classes_teste = train_test_split(notas.reshape(-1,1),
                                                                            classes.reshape(-1,1), 
                                                                            test_size=0.2, 
                                                                            random_state=42)
```

```python
modelo = DecisionTreeClassifier() # crio meu modelo de árvore de decisão
modelo.fit(notas_treino,classes_treino) # Treinamento do modelo de árvore de decisão
```

```python
previsoes = modelo.predict(notas_teste)  # dentro de modelo predict  é colocado as notas de teste
previsoes
```

```python
notas_teste
```

```python
# teste de nota individual
nota = np.array([4,7]).reshape(-1,1) 
resultado = modelo.predict(nota)
resultado
```

```python
fig = plt.figure(figsize=(10,8)) # cria uma figura de tamanho 10 x 8
# tree plot plota a árvore de decisão
# modelo é o modelo criado do algoritmo de aprendizado supervisionado
tree.plot_tree(modelo,feature_names=notas.tolist(),class_names=classes.tolist(),filled=True)
plt.title('Árvore de decisão para classificar uma bebida \n'
          'de acordo com a nota',fontsize=18)
plt.show()
```

```python
# plota box plot com a biblioteca seaborn 
# x é o eixo x que indica as classes : boa, ruim ou péssima
# y indica as notas
plt.figure(figsize=(10,8))


plt.title('Box plot das notas e classes das bebidas')

sns.boxplot(x = classes, y = notas)
plt.grid()
```

```python
# gráfico de barras
plt.figure(figsize=(10,8))
sns.barplot(y=notas, x = classes)
plt.title('Notas de bebidas',fontsize=20)
plt.xlabel('Classificação',fontsize=16)
plt.ylabel('Notas',fontsize=16)
```

```python
# métrica do sistema
acuracy = accuracy_score(classes_teste,previsoes)
acuracy
```

```python
precisao = precision_score(classes_teste,previsoes,pos_label='bom')
precisao
```

```python
recall = recall_score(classes_teste,previsoes,pos_label='ruim')
recall
```

```python
f1score = f1_score(classes_teste,previsoes,pos_label='bom')
f1score
```
